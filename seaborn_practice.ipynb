```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("sns.csv")
df.head()
```

```python
plt.plot(df["week"], df["attendance_rate"], marker='o')
plt.title("Weekly Attendance Rate")
plt.xlabel("Week")
plt.ylabel("Attendance Rate (%)")
plt.show()
```

```python
sb.lineplot(data=df,x = "week",y ='attendance_rate',errorbar=None,hue="gender")
```

```python

sb.scatterplot(data=df, x = "study_hours" , y= "test_score", hue='gender')
```

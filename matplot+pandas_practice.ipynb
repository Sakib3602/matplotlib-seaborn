```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("Practice Day (1).csv", on_bad_lines='skip')
df
```

```python
# df["Total Marks"] = df['Marks1'] + df['Marks2'] + df['Marks3']
df["Total Marks"] = df.iloc[::,5:8].sum(axis=1)
df
```

```python
df["Discounted_Price"] = df["Total Marks"] * 0.9
df
```

```python
df["EmployeeID"].duplicated().sum()
```

```python
df.drop_duplicates(subset="EmployeeID",inplace=True)
```

```python
df
```

```python
df["Age"].isna()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean()).round()
df["Salary"] = df["Salary"].fillna(df["Salary"].mean()).round(2)
```

```python
df
```

```python
plt.hist(df["Age"],color="#F37626",edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Histrogram")
```

```python
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Scatterplot")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Practice Day (1).csv",on_bad_lines='skip')
df
```

```python
df["Discounted_Price"] = df["Price"] * 0.9
df
```

```python
df.to_csv("discounted_products.csv")
```

```python
df.duplicated("EmployeeID").sum()
```

```python
df.drop_duplicates("EmployeeID",inplace=True)
```

```python
df.duplicated("EmployeeID").sum()
```

```python
df.tail()
```

```python
df["Age"].isna().sum()
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Age"] = df["Age"].fillna(df["Age"].mean()).round(2)
df
```

```python
df["Age"].isna().sum()
```

```python
df["Salary"].isna().sum()
```

```python
df.dropna(subset="Salary",inplace=True)
```

```python
df["Salary"].isna().sum()
```

```python
df
```

```python

df["Total Marks"] = df.iloc[:, 5:8].sum(axis=1)
df
```

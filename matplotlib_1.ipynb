```python
import matplotlib.pyplot as plt

hours = [1, 3, 5, 7, 8, 2, 1, 5]
hours2 = [10,2,3,2,12,3,8,9]
days = [1, 2, 3, 4, 5, 6, 7, 8]

plt.plot(days,hours, label="Sleeper 1", color="pink",)
plt.plot(days,hours2, label="Sleeper 2",linestyle='dotted',marker="o")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Sleep Time")

plt.legend()
```

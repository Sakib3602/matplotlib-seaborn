```python
df["Salary"].isna().sum()
```

```python
# Fill missing Salary with department-wise median
df['Salary'] = df['Salary'].fillna(df.groupby('Department')['Salary'].transform('median'))
df['Salary'].isna().sum()
```

```python
df.head()
```

```python

x1 = df['Salary'].where((df['ExperienceYears'] > 12) & ( df["ProjectCount"] > 5) ).sum()
x1

```

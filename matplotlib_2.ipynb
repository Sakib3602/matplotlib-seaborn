```python
import pandas as pd
from datetime import datetime
import random

# Create year range from 2000 to current year
years = list(range(2000, datetime.now().year + 1))

# Generate random number of students for each department
cse_students = [random.randint(50, 200) for _ in years]
bba_students = [random.randint(40, 180) for _ in years]

# Create DataFrame
df = pd.DataFrame({
    "Year": years,
    "CSE Students": cse_students,
    "BBA Students": bba_students
})

print(df)
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Year"], df["CSE Students"],label="CSE",marker="o")
plt.plot(df["Year"], df["BBA Students"],label="BBA",marker='o')
plt.xlabel("YEARS")
plt.ylabel("STUDENT ENROLLMENTS")
plt.title("FROM 2000 to 2025 STUDENT ON CSE AND BBA")
plt.legend()
plt.grid()
```

# ***Scatter plot***

```python
import random

# Sample student names
students = [f"Student {i}" for i in range(1, 11)]  # 10 students

# Generate random data
study_hours = [random.uniform(2, 8) for _ in students]   # 28 hours
sleep_hours = [random.uniform(5, 9) for _ in students]   # 59 hours
chilling_hours = [random.uniform(1, 5) for _ in students] # 15 hours
iq_scores = [random.randint(90, 150) for _ in students]  # IQ between 90150

# Create DataFrame
df = pd.DataFrame({
    "Student": students,
    "Study Hours": [round(h, 1) for h in study_hours],
    "Sleep Hours": [round(h, 1) for h in sleep_hours],
    "Chilling Hours": [round(h, 1) for h in chilling_hours],
    "IQ Score": iq_scores
})

print(df)
```

```python
plt.scatter(df["Study Hours"], df['IQ Score'],label="study hour & IQ")
plt.xlabel("Hours")
plt.ylabel("IQ lavel")
plt.title("Relation between study hour and Iq score")

plt.legend()
```

```python
plt.scatter(df["Sleep Hours"], df['IQ Score'],label="Sleep Hours & IQ")
plt.xlabel("Sleep Hours")
plt.ylabel("IQ lavel")
plt.title("Relation between sleep hour and Iq score")

plt.legend()
```

```python
plt.scatter(df["Chilling Hours"], df['IQ Score'],label="Chilling Hours & IQ")
plt.xlabel("Chilling Hours")
plt.ylabel("IQ lavel")
plt.title("Relation between Chilling hour and Iq score")

plt.legend()
```
